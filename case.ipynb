```python
import pandas as pd
import numpy as np

spend_cols = ['spend_1d', 'spend_3d', 'spend_7d', 'spend_14d', 'spend_30d']
days_map = {'spend_1d':1, 'spend_3d':3, 'spend_7d':7, 'spend_14d':14, 'spend_30d':30}

def get_last_spend_day(row):
    last_day = 0
    prev_val = 0
    for col in spend_cols:
        val = row[col]
        if val > prev_val:
            last_day = days_map[col]
            prev_val = val
    return last_day

firstpackmerged['last_spend_day'] = firstpackmerged.apply(get_last_spend_day, axis=1)
firstpackmerged['is_payer'] = firstpackmerged['amount'] > 0

segment_cols = ['gender', 'age_group', 'country_group', 'os', 'browser']  # traffic_type видалено

agg = firstpackmerged.groupby(segment_cols + ['split_group']).agg(
    users=('id_user', 'nunique'),
    payers=('is_payer', 'sum'),
    revenue=('amount', 'sum'),
    avg_retention=('last_spend_day', 'mean')
).reset_index()

agg['CR_to_paid'] = agg['payers'] / agg['users']
agg['ARPU'] = agg['revenue'] / agg['users']
agg['ARPPU'] = agg['revenue'] / agg['payers'].replace(0, np.nan)

control = agg[agg['split_group'] == 0].copy().rename(columns={
    'CR_to_paid': 'control_CR_to_paid',
    'ARPU': 'control_ARPU',
    'ARPPU': 'control_ARPPU',
    'avg_retention': 'control_retention'
})

test = agg[agg['split_group'] == 1].copy().rename(columns={
    'CR_to_paid': 'test_CR_to_paid',
    'ARPU': 'test_ARPU',
    'ARPPU': 'test_ARPPU',
    'avg_retention': 'test_retention'
})

control['user_count'] = control['users']
test['user_count'] = test['users']

comparison = test.merge(control, on=segment_cols, suffixes=('_test', '_control'))

comparison['user_count'] = comparison['user_count_test'] + comparison['user_count_control']


comparison['score_CR'] = (comparison['test_CR_to_paid'] > comparison['control_CR_to_paid']).astype(int)
comparison['score_ARPU'] = (comparison['test_ARPU'] > comparison['control_ARPU']).astype(int)
comparison['score_ARPPU'] = (comparison['test_ARPPU'] > comparison['control_ARPPU']).astype(int)
comparison['score_retention'] = (comparison['test_retention'] > comparison['control_retention']).astype(int)

comparison['score'] = comparison[['score_CR', 'score_ARPU', 'score_ARPPU', 'score_retention']].sum(axis=1)

def assign_action(score):
    if score >= 3:
        return "в тестову"
    elif score <= 1:
        return "в контрольну"
    else:
        return "не чіпати"

comparison['action'] = comparison['score'].apply(assign_action)

final_result = comparison[segment_cols + [
    'test_CR_to_paid', 'control_CR_to_paid',
    'test_ARPU', 'control_ARPU',
    'test_ARPPU', 'control_ARPPU',
    'test_retention', 'control_retention',
    'user_count',
    'score', 'action'
]].sort_values(by='user_count', ascending=False)
```

```python
segment_cols = ['gender', 'age_group', 'country_group', 'os', 'browser']

low_score_segments = final_result[final_result['score'].isin([0, 1])][segment_cols].drop_duplicates()
high_score_segments = final_result[final_result['score'].isin([3, 4])][segment_cols].drop_duplicates()

low_score_segments.to_csv('low_score_segments.csv', index=False)
high_score_segments.to_csv('high_score_segments.csv', index=False)
```
